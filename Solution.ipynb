```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")
df=df.drop('id',axis=1)
df.head()
```

```python
from sklearn import preprocessing
le = preprocessing. LabelEncoder()
```

```python
df_cat=df.select_dtypes(include='object')
df_num=df.select_dtypes(include='float64')
cols=list(df_cat.columns)
cols
```

```python
for col in cols:
    impute_col = col
    le.fit(df.loc[:, impute_col])
    dict1=dict(zip(le.classes_, le.transform(le.classes_)))
    df[impute_col]=le.transform(df[impute_col])
    df[impute_col]= df[impute_col].map(lambda x: np.nan if x == len(dict1)-1 else x)
```

```python
# df.head(200)
```

```python
imputer = IterativeImputer(random_state=100)
```

```python
for col in cols:
    impute_col = col
    df_train = df.loc[:, [impute_col]]
    imputer.fit(df_train)
    df_imputed=imputer.transform(df_train)
    df_imputed[:,0].round()
    df.loc[:,[impute_col]]=df_imputed[:,0].round()
```

### Creating the pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])
```

```python
df_num_tr = my_pipeline.fit_transform(new_df)
df_num_tr
```

```python
df_num_tr.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
    # Add other hyperparameters you want to tune
}
grid_search = GridSearchCV(RandomForestRegressor(), param_grid, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)

grid_search.fit(df_num_tr, df_lables)
best_params = grid_search.best_params_
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor


estimators = [
    ('rf', LinearRegression()),
    ('svr', SVR()),
    ('dt', DecisionTreeRegressor()),
    ('knn', KNeighborsRegressor()),
    ('xgb',XGBRegressor(**best_params) ),
]
```

```python
# model = RandomForestRegressor()
# model=DecisionTreeRegressor()
# model = LinearRegression()
model = StackingRegressor(estimators=estimators, final_estimator=RandomForestRegressor(**best_params))
```

```python
model.fit(df_num_tr, df_lables)
```

### Evaluating the model

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
df_predictions = model.predict(df_num_tr)
mse = mean_squared_error(df_lables, df_predictions)
mae=mean_absolute_error(df_lables, df_predictions)
rmse = np.sqrt(mse)
mae
```

```python
rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, df_num_tr, df_lables, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
def print_scores(scores):
    print("Scores:", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
```
